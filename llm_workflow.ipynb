```python
from langchain_ollama import ChatOllama
llm = ChatOllama(model="llama3.1:8b")
```

```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
class llmstate(TypedDict):
    q:str
    a:str  
    
```

```python
def llmqa(state: llmstate)-> llmstate:
    q=state['q']

    prompt= f"Answer the following question {q}"

    a= llm.invoke(prompt)

    state['a']=a

    return state
    
```

```python
graph= StateGraph(llmstate)

graph.add_node('llmqa', llmqa)

graph.add_edge(START, 'llmqa')
graph.add_edge('llmqa', END)

workflow=graph.compile()
```

```python
initial= {'q': 'speed of light'}

final= workflow.invoke(initial)

print(final['a'])
```
